```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Estos son los primeros resultados obtenidos de la etapa 1 del dise√±o experimental.
Estos datos seran utilizados en un primer ensayo de entrenamiento a nivel de expresion de transcritos para el ML-No supervisado.

```python
#Grouped bar chart: 'Isoformas vs trancritos √∫nicos' (DATASET 1) 

#labels = ['ApCf48','ZmCg24','MdAa18','AtBc24','AtSs30','SlSs26','PnPc24']
labels = ['ApCf48 [~30]','ZmCg24 [~0.8]','MdAa18 [~12]','AtBc24 [~12]','AtSs30 [~21.8]','SlSs26 [~22.7]','PnPc24 [~27.5]']
isoforms = [151346,3201,19145,33613,33930,47452,120244] 
unique_isoform = [80293,3162,18267,29858,29974,41475,53789]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(20,10))

rects1 = ax.bar(x - width/2, isoforms, width, label='Isoformas')
rects2 = ax.bar(x + width/2, unique_isoform, width, label='T.√∫nicos')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('N√∫mero de transcritos',fontsize=20, fontweight='bold')
ax.set_title('Isoformas y transcritos √∫nicos',fontsize=30, fontweight='bold')
ax.set_xticks(x)
ax.set_xlabel('Transcriptomas',fontsize=20, fontweight='bold')
ax.set_xticklabels(labels,fontsize=18)
ax.legend(fontsize=16)


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=14, fontweight='bold')
        
autolabel(rects1)
autolabel(rects2)
#fig.tight_layout()
#plt.show()
```
