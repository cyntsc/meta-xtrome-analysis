```python

import pandas as pd  # This is always assumed but is included here as an introduction. import numpy as np import matplotlib.pyplot as plt np.random.seed(0)
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Contigs and Unique contigs in the trasncriptomes

dataFrom = ['Apple','Malus','Arabi1','Arabi2','Solanum','Pipper']
Contigs = [151346,19145,33613,33930,47452,120244] 
UniContigs = [80293,18267,29858,29974,41475,53789]
plt.figure(figsize=(20,10))
line_chart1 = plt.plot(dataFrom, Contigs,'--')
line_chart2 = plt.plot(dataFrom, UniContigs,':',linewidth=4)
plt.plot(dataFrom, UniContigs, '*', markersize=20, color='red')

plt.xticks(fontsize=20, fontweight='bold',rotation=45)
plt.yticks(fontsize=20, fontweight='bold')
plt.title('Contigs and Unicontigs by transcriptome',fontsize=30, fontweight='bold')
plt.xlabel('Transcriptomes',fontsize=25, fontweight='bold')
plt.ylabel('Total contigs',fontsize=25, fontweight='bold')
plt.legend(['Contigs', 'Unicontigs'], fontsize=25, loc=4)
```

```python
# N50 Metrics for assessment fo the assembly

import matplotlib.pyplot as plt2

dataFrom = ['Apple','Malus','Arabi1','Arabi2','Solanum','Pipper']
N50 = [1799,649,1358,1693,1436,1283]
Maxlng_Contig = [12417,5520,13274,13226,10392,9993]
#Se redondearon los valores del tamao promedio del contig en cada transcriptoma por fines de visualizacin
Avg_Contig_Size = [1189,568,936,1125,943,953]
plt2.figure(figsize=(20,10))
line_chart1 = plt2.plot(dataFrom, N50,'--')
line_chart2 = plt2.plot(dataFrom, Maxlng_Contig,'*')
line_chart3 = plt2.plot(dataFrom, Avg_Contig_Size,'--')
plt2.plot(dataFrom, Maxlng_Contig, '*', markersize=20, color='black')
plt2.xticks(fontsize=20, fontweight='bold',rotation=45)
plt2.yticks(fontsize=20, fontweight='bold')
plt2.title('Assessment of the assembly (min_len = 250)',fontsize=30, fontweight='bold')
plt2.ylabel('Contig length',fontsize=25, fontweight='bold')
plt2.xlabel('Transcriptomes',fontsize=25, fontweight='bold')
plt2.legend(['N50-Contig','Max-Contig-Lng','Avg_Contig_Size'], fontsize=25, loc=4)
```
